```python
genre_counts = data['Genre'].value_counts().reset_index()
genre_counts.columns = ['Genre', 'Count']
top_n_genres = genre_counts.head(10)
top_n_genres

```

```python
plt.figure(figsize =(4,4))
plt.pie(top_n_genres['Count'], labels = top_n_genres['Genre'], autopct = '%1.1f%%')
plt.title('Top 10 Genre of the movies watched ')
plt.show()
```

```python
dir_rating = data.groupby('Director')['Rating'].mean().reset_index()
dir_rating.columns = ['Director', 'Average Rating']
dir_rating = dir_rating.sort_values(by = 'Average Rating', ascending = False)
top_directors = dir_rating.head(10)
top_directors
```

```python
plt.figure(figsize = (10,6))
sns.barplot(x = 'Average Rating', y = 'Director', data = top_directors)
#plt.xticks(rotation = 90)
plt.xlabel('Director')
plt.ylabel('Average Rating')
```

```python
actor_rating = data['Actor 1'].value_counts().reset_index()
actor_rating.columns = ['Actor', 'Average Rating']
actor_rating = actor_rating.sort_values(by = 'Average Rating', ascending = False)
top_actors = actor_rating.head(10)
top_actors
```

```python
plt.figure(figsize = (10,6))
sns.barplot(x = 'Average Rating', y = 'Actor', data = top_actors)
#plt.xticks(rotation = 90)
plt.xlabel('Actor')
plt.ylabel('Average Rating')
```

```python
year_rating = data.groupby('Year')['Rating'].mean().reset_index()
year_rating.columns = ['Year','Average Rating']
year_rating = year_rating.sort_values(by = 'Average Rating', ascending = False)
year_rating= year_rating.head(10)

```

```python
plt.figure(figsize = (10,6))
sns.barplot(x = 'Year', y = 'Average Rating', data = year_rating)

#plt.xticks(rotation = 90)
plt.xlabel('Year')
plt.ylabel('Average Rating')
#
```

```python
from sklearn.preprocessing import LabelEncoder
categorical_columns = data.select_dtypes(include=['object']).columns
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le
#LabelEncoder = LabelEncoder()
#data['Genre'] = LabelEncoder.fit_transform(data['Genre'])
#data['Actor1']
```

```python
features = data.drop(['Rating'], axis=1)
target = data['Rating']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
X_test, X_train, y_test, y_train = train_test_split(features, target, test_size=0.2, random_state=42)
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print (y_pred)
```

```python
from sklearn.metrics import  r2_score
r2 = r2_score(y_test, y_pred)
accuracy = r2 * 100
print("Accuracy:", accuracy)
```
